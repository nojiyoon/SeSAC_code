```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.enable_eager_execution()
tf.__version__
```

```python
X = np.array([1, 2, 3], dtype="float32")
Y = np.array([2, 2.5, 3.5], dtype="float32")
```

```python
W = tf.Variable([2], dtype="float32")
b = tf.Variable([1], dtype="float32")
```

```python
W
```

```python
b
```

- tensorflow는 기본적으로 W, b 값을 출력 불가!
- W, b 값 출력 하고싶다면 맨 처음에 tf.enable_eager_execution() 선언해야됨

```python
learning_rate = 0.1
```

```python
hypothesis = W * X + b
print("hypothesis:", hypothesis)    # 예측값

print("hypothesis - y:", (hypothesis - Y))  # 예측값 - 진짜값
```

```python
hypothesis = W * X + b
print("hypothesis:", hypothesis)    # 예측값

print("hypothesis - y:", (hypothesis - Y))  # 예측값 - 진짜값
print("tf.square(hypothesis - Y):", tf.square(hypothesis - Y)) # (예측값 - 진짜값)의 제곱
```

```python
hypothesis = W * X + b
print("hypothesis:", hypothesis)    # 예측값

print("hypothesis - y:", (hypothesis - Y))  # 예측값 - 진짜값
print("tf.square(hypothesis - Y):", tf.square(hypothesis - Y)) # (예측값 - 진짜값)의 제곱
print("tf.reduce_mean(tf.square(hypothesis - Y)):", tf.reduce_mean(tf.square(hypothesis - Y))) # (예측값 - 진짜값)의 평균
```

```python
hypothesis = W * X + b
print("hypothesis:", hypothesis)    # 예측값

# print("hypothesis - y:", (hypothesis - Y))  # 예측값 - 진짜값
# print("tf.square(hypothesis - Y):", tf.square(hypothesis - Y)) # (예측값 - 진짜값)의 제곱
# print("tf.reduce_mean(hypothesis - Y):", tf.reduce_mean(hypothesis - Y)) # (예측값 - 진짜값)의 평균

cost = tf.reduce_mean( tf.square(hypothesis - Y) )
print("cost = ", cost)
```

```python
hypothesis = W * X + b
print("hypothesis:", hypothesis)    # 예측값

cost = tf.reduce_mean( tf.square(hypothesis - Y) )
print("cost = ", cost)
tape = tf.GradientTape()

W_grad, b_grad = tape.gradient(cost, [W, b])
print("W_grad", W_grad, ', b_grad', b_grad)
```

- 에러는 안 나지만 기울기 계산 안됨

```python
with tf.GradientTape() as tape:

  hypothesis = W * X + b
  print("hypothesis:", hypothesis)    # 예측값

  cost = tf.reduce_mean( tf.square(hypothesis - Y) )
  print("cost = ", cost)

W_grad, b_grad = tape.gradient(cost, [W, b])
print("W_grad", W_grad, ', b_grad', b_grad)
```

- tape가 맨 처음에 나오고 그 안에 hypothesis, cost가 있어야 됨

```python
with tf.GradientTape() as tape:
  hypothesis = W * X + b
  print("hypothesis:", hypothesis)
  cost = tf.reduce_mean(tf.square(hypothesis - Y))
  print("cost:", cost)

W_grad, b_grad = tape.gradient(cost, [W, b])
print("W_grad:", W_grad, ", b_grad:", b_grad)

W.assign_sub(W_grad * learning_rate)
b.assign_sub(b_grad * learning_rate)
```

```python
# 반복해서 W, b를 업데이트
for i in range(1000):
  # 기울기 계산할 객체
  # 프로그램 끝까지 살아있음
  with tf.GradientTape() as tape:
    hypothesis = W * X + b
    print("hypothesis:", hypothesis)
    cost = tf.reduce_mean(tf.square(hypothesis - Y))
    print("cost:", cost)

  W_grad, b_grad = tape.gradient(cost, [W, b])
  print("W_grad:", W_grad, ", b_grad:", b_grad)

  W.assign_sub(W_grad * learning_rate)
  b.assign_sub(b_grad * learning_rate)

  print("=" * 100)
```
