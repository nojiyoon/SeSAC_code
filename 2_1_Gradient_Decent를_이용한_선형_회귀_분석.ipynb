```python
# Gradient Decent 편미분 공식에 의해서 w_gred 계산
w_gred = 1/3 * ( np.dot( (hypothesis-y), X.T ) )
```

```python
for i in range(100):
  # 예측값을 hypothesis에 대입
  hypothesis = X * w + b
  print("hypothesis:", hypothesis)

  # 예측값 - 진짜값의 제곱의 합 cost를 계산
  MSE = np.sum((hypothesis-y)**2)
  print("MSE:", MSE)
  
  # w의 수정값
  w_gred = 1/3 * np.dot((hypothesis-y), X.T)
  print("w_gred:", w_gred)
  
  # b의 수정값
  b_gred = 1/3 * np.sum(hypothesis-y)
  print("b_gred:", b_gred)
  
  # w 업데이트
  w = w - w_gred
  print("w:", w)
  
  # b 업데이트
  b = b - b_gred
  print("b:", b)

  print("=" * 100)
```

```python
w = 2
b = 1
learning_rate = 0.1
```

```python
for i in range(100):
  # 예측값을 hypothesis에 대입
  hypothesis = X * w + b
  print("hypothesis:", hypothesis)

  # 예측값 - 진짜값의 제곱의 합 cost를 계산
  MSE = np.sum((hypothesis-y)**2)
  print("MSE:", MSE)
  
  # w의 수정값
  w_gred = 1/3 * np.dot((hypothesis-y), X.T)
  print("w_gred:", w_gred)
  
  # b의 수정값
  b_gred = 1/3 * np.sum(hypothesis-y)
  print("b_gred:", b_gred)
  
  # w 업데이트
  w = w - (learning_rate * w_gred)
  print("w:", w)
  
  # b 업데이트
  b = b - (learning_rate * b_gred)
  print("b:", b)

  print("=" * 100)
```
